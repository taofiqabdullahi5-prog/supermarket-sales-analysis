```python
import pandas as pd

!git clone https://github.com/taofiqabdullahi5-prog/supermarket-sales-analysis.git

df = pd.read_csv("supermarket-sales-analysis/data/supermarket_sales.csv")

# Top 5 rows
print("Top 5 rows:")
print(df.head())

# Number of columns and info
print("\nNumber of columns and info:")
print(df.info())

# Check for duplicates
print("\n✅ DUPLICATES:")
print(df[df.duplicated(keep=False)])

# Check for negative bill
print("\n✅ NEGATIVE BILL:")
print(df[df['total_bill'] < 0])

# Check unique values in 'sex' column
print("\n✅ SEX CATEGORY UNIQUENESS:")
print(df['sex'].unique())

# Check for negative tips
negative_tips = df[df['tip'] < 0]
print("\n✅ Negative Tips:")
print(negative_tips)


# Check for highest total bill
Highest = df['total_bill'].idxmax()
print("\n✅Highest total bill")
print(df.loc[Highest]) # Corrected: used variable 'Highest' directly

# Check for inconsistent value
print("====INCONSISTENT CHECK====")
col_to_check = ['sex', 'smoker', 'day', 'time']
for col in col_to_check:
    print(f"column: {col}")
    print(f"unique value: {df[col].unique()}")
    print(f"value counts: {df[col].value_counts()}")
    print("-" * 40)

# Total revenue of the restaurant
print("\n✅ Total revenue")
print(df['total_bill'].sum())


# Average spending
print("\n✅ Average Customer Spending")
Avg = df['total_bill'].mean()
print(f"{Avg:.2f}")

# percentage tip
print("\n✅ Average tip percentage")
Perc = (df['tip'] / df['total_bill'] * 100).mean()
print(f"{Perc:.2f}%")

print(df['tip'].mean())



import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

# Scatter plot
plt.scatter(df['total_bill'], df['tip'], alpha=0.6)

# Fit straight line: y = mx + b
m, b = np.polyfit(df['total_bill'], df['tip'], 1) # 1 = degree 1 = straight line
plt.plot(df['total_bill'], m*df['total_bill'] + b, color='red', linewidth=2, label=f'y = {m:.2f}x + {b:.2f}')

plt.title("Tip vs Total Bill with Trend Line")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Equation: Tip = {m:.2f} * Total_Bill + {b:.2f}")
```

🎯*Interpretation of the scatter plot*



The scatter plot reveals a positive linear relationship between total bill and tip amounts: higher bills generally lead to higher tips.

*The regression model (Tip = 0.11 × Total Bill + 0.92)*

 shows that for every additional dollar spent, the expected tip rises by about (0.11 dollar) on average.
While total bill is a significant predictor, substantial scatter around the line indicates other factors (e.g., service quality, party size, customer habits) also influence tipping behavior. The intercept ($0.92) is a mathematical artifact with limited real-world meaning.





🎯*Business insights*



The analysis indicates that increasing customer spending is likely to increase total tips. Management could consider strategies such as meal bundles, premium menu offerings, or upselling techniques to encourage higher spending. However, because the relationship is not perfect, maintaining high service quality remains important, as customer experience may also affect tipping behavior.
